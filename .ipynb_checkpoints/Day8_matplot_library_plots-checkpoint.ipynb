```python
x=list(range(10))
y=list(range(10,20))
# plt.plot([1,2,3,4],[1,8,27,64],c='blue',lw=5,label='line',ls='-.')
plt.plot([1,2,3,4],[1,8,27,64],'-bo',alpha=0.5,lw=5,label='line')
plt.xlabel('x-label')
plt.ylabel('y-label')
#plt.xlim(-7,20)
#plt.ylim(-3,50)
plt.legend(loc='upper center')
plt.title('x versus y')
plt.scatter(x,y,c='g',marker='*',s=150,alpha=0.7)
```

```python
state = ['Delhi','UP','HP','MP']
```

```python
population = [1000,2000,300,1400]
```

```python
income = [100,50,20,80]
plt.bar(state,population,color='blue')  #vertical
plt.bar(range(len(state)),income,color='red')
# plt.barh(state,population)   #horizontal
plt.xticks(range(len(state)),['A','B','C','D'])
#plt.yticks(range(len(state)),['A','B','C','D'])
```

```python
income = [100,50,20,80]
#plt.bar(state,population,color='blue')  #vertical
plt.bar(np.array(range(len(state)))-0.2,population,color='red',width=0.3,label='0')
plt.bar(np.array(range(len(state)))+0.2,income,color='blue',width=0.3,label='1')
# plt.barh(state,population)   #horizontal
plt.xticks(range(len(state)),['A','B','C','D'])
#plt.yticks(range(len(state)),['A','B','C','D'])
plt.legend(loc='upper right')
```
