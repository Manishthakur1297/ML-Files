```python
import numpy as np
import pandas as pd
from sklearn import naive_bayes
from sklearn import cross_validation
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('smsspam',names=['status','messages'],delimiter='\t')
```

```python
df.head()
```

```python
df['status'][df['status']=='spam'] = 1
df['status'][df['status']=='ham'] = 0
y = df['status'].values
X = df['messages']
X.head()
```

```python
y
```

```python
y = y.astype('int64')
```

```python
y
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
X_cv = cv.fit_transform(X)
```

```python
X_cv.toarray()
```

```python
X_cv.shape
```

```python
X_train,X_test,y_train,y_test = cross_validation.train_test_split(X_cv,y,test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
nbm = naive_bayes.MultinomialNB()
```

```python
nbm.fit(X_train,y_train)
```

```python
nbm.score(X_test,y_test)
```
